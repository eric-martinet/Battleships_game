```python
class Player:
    def __init__(self, name):
        self.name = name
        print(f"Player {self.name} created!")
        self.create_grids()
        self.create_ships()

    
    def create_grids(self):
        self.grid=[['-' for i in range(10)] for j in range(10)]
        self.displayed_grid=[['-' for i in range(10)] for j in range(10)]
        self.grid_filled_coordinates = []
        
    def display_grid(self):
        print("\n" + f"Board of {self.name}:" + "\n")
        print("\n".join(["".join(self.displayed_grid[i]) for i in range(10)]))
        
    def create_ships(self):
        self.carrier = Ship("Carrier", 5)
        self.battleship = Ship("Battleship", 4)
        self.cruiser = Ship("Cruiser", 3)
        self.submarine = Ship("Submarine", 3)
        self.destroyer = Ship("Destroyer", 3)
        self.ships = [self.carrier, self.battleship, self.cruiser, self.submarine, self.destroyer]
        
    def update_grid(self, ship, coordinates):
        for coord in coordinates:
            self.grid[coord[0]][coord[1]] = ship
        self.update_displayed_grid()
        
    def update_displayed_grid(self):
        for i in range(10):
            for j in range(10):
                if type(self.grid[i][j])== Ship:
                    self.displayed_grid[i][j] = "X"
        
    def place_ships(self):
        for ship in self.ships:
            if self.place_ship(ship):
                self.update_grid(ship, ship.coordinates)
            self.display_grid()
    
    def place_ship(self, ship):
        print(f"\nPlacing {ship.model} of size {ship.size}!")
        user_input = input("Please enter coordinates (0-9) (0-9) (right or bottom): ")
        coordinates_input = user_input.split(" ")
        print(coordinates_input)
        i = int(coordinates_input[0])
        j = int(coordinates_input[1])
        d = coordinates_input[2]
        ship.update_coordinates((i,j), d)
        if not self.check_place_ship(ship):
                print("Please retry")
                self.place_ship(ship)
        else:
            self.grid_filled_coordinates.extend([tuple(coord) for coord in ship.coordinates])
        return True

            
    def check_place_ship(self, ship):
        if ship.origin[0] > 9:
            print("Wrong input: first digit (line) must be less than 10.")
            return False
        if ship.origin[1] > 9:
            print("Wrong input: first digit (column) must be less than 10.")
            return False
        for coord in ship.coordinates:
            if tuple(coord) not in base_grid:
                print(tuple(coord))
                print("The ship does not fit on the board.")
                return False
            if tuple(coord) in self.grid_filled_coordinates:
                print(tuple(coord))
                print(self.grid_filled_coordinates)
                print("Overlap.")
                return False
        return True
```

```python
class Ship:
    def __init__(self, model, size):
        self.model = model
        self.size = size
        self.life = size
        self.origin = []
        self.coordinates=[]
        
    def update_coordinates(self, origin, direction):
        self.coordinates=[]
        if direction == "bottom" or direction == "b":
            for i in range(self.size):
                self.coordinates.append([list(origin)[0]+i, list(origin)[1]])
        else:
            for i in range(self.size):
                self.coordinates.append([list(origin)[0], list(origin)[1]+i])
        self.update_origin(self.coordinates)
    
    def update_origin(self, coordinates):
        self.origin = coordinates[0]
```
